# Base de datos propinas

* Objeto a estudiar cantidad de propinas.
    - `tip`: Cantidad de la propina
* Atributos de nuestra base de datos:
    - `sex`: Sexo de camarero/a. String.
    - `day`: Dia del evento.
    - `time`: Almuerzo o cena.
    - `smoker`: Zona fumadores o no. 
    - `total_bill`: Total de la cuenta.
    - `size`: Tama√±o de la mesa.

```python
%matplotlib inline
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as statm
import statsmodels.formula.api as frm
plt.style.use('seaborn')
```

```python
# vamos a importar nuestro csv
df = pd.read_csv('./tips.csv')
```

```python
df['sex'].replace(['Female', 'Male'], [1, 0], inplace=True)
```

```python
df.head()
```

```python
df['smoker'].replace(['Yes', 'No'], [1, 0], inplace=True )
```

```python
df.head()
```

```python
df['time'].replace(['Lunch', 'Dinner'], [1, 0], inplace=True)
```

```python
df.head()
```

```python
df.assign(weekend = df.day.map({'Sat': 1, 'Sun': 1, 'Thur': 0, 'Fri': 0}))
```
